# ** Inheritance (Generalization)**
Child class inherits properties from Parent class.

Doctor is a Person
Student is a Person

inheritance1.PNG


```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Doctor(Person):
    def __init__(self, name, age, specialty):
        super().__init__(name, age)
        self.specialty = specialty

class Student(Person):
    def __init__(self, name, age, roll_no):
        super().__init__(name, age)
        self.roll_no = roll_no

d1 = Doctor("Ali", 40, "Cardiology")
print(d1.name, d1.specialty)
```

**ASSOCIATION (HAS-A Relationship)**
Concept

One class is connected to another class.

Customer HAS Orders.

image.png

```python
class Customer:
    def __init__(self, name):
        self.name = name

class Order:
    def __init__(self, order_id, customer):
        self.order_id = order_id
        self.customer = customer
```

# **MULTIPLE INHERITANCE**

One class inherits from multiple parents.

TeachingAssistant IS-A Student AND Employee.

image.png

```python
class Student:
    def __init__(self, roll_no, name, cgpa):
        self.roll_no = roll_no
        self.name = name
        self.cgpa = cgpa

    def enroll_course(self, course):
        print(f"{self.name} enrolled in {course}")

    def get_cgpa(self):
        return self.cgpa


class Employee:
    def __init__(self, employee_id, salary):
        self.employee_id = employee_id
        self.salary = salary

    def get_salary(self):
        return self.salary

    def calculate_bonus(self):
        return self.salary * 0.10


class TeachingAssistant(Student, Employee):
    def __init__(self, roll_no, name, cgpa,
                 employee_id, salary, assigned_course):

        Student.__init__(self, roll_no, name, cgpa)
        Employee.__init__(self, employee_id, salary)

        self.assigned_course = assigned_course

    def take_lab(self):
        print(f"{self.name} is taking lab of {self.assigned_course}")

    def show_profile(self):
        print(f"TA: {self.name}, CGPA: {self.cgpa}, Salary: {self.salary}")
```

# **MULTILEVEL INHERITANCE**

Grandparent → Parent → Child

**Example:**
Person → Employee → Professor

image.png

```python
class Employee(Person):
    def __init__(self, name, age, cnic, employee_id, salary):
        super().__init__(name, age, cnic)
        self.employee_id = employee_id
        self.salary = salary

    def calculate_salary(self):
        return self.salary

class Doctor(Employee):
    def __init__(self, name, age, cnic, employee_id, salary, specialization):
        super().__init__(name, age, cnic, employee_id, salary)
        self.specialization = specialization

    def diagnose(self):
        print("Diagnosing patient")
class Researcher:
    def __init__(self, research_field):
        self.research_field = research_field

    def publish_paper(self):
        print("Research paper published")
class SurgeonScientist(Doctor, Researcher):
    def __init__(self, name, age, cnic, employee_id, salary,
                 specialization, research_field):
        Doctor.__init__(self, name, age, cnic, employee_id, salary, specialization)
        Researcher.__init__(self, research_field)

    def perform_surgery(self):
        print("Performing surgery")
```

# **ONE EXAMPLE THAT COVER ALL OOP CONCEPT **
This UML diagram represents a simple online shopping system. The Order class is the main part that contains one or more OrderItem objects (composition), meaning items cannot exist without the order. The order uses a Payment method to complete the purchase, but Payment is an abstract class, so we cannot create it directly. Instead, CreditCardPayment and JazzCashPayment inherit from Payment and provide their own implementation of the pay() method (this is polymorphism). The Invoice class implements the Printable interface, which means it must define the print_receipt() method. Overall, this design shows abstraction, inheritance, polymorphism, interface implementation, composition, and dependency in a simple e-commerce system.

image.png

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass


class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")


class JazzCashPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using JazzCash")


class OrderItem:
    def __init__(self, name, quantity, price):
        self.name = name
        self.quantity = quantity
        self.price = price

    def total(self):
        return self.quantity * self.price


class Printable:
    def print_receipt(self):
        pass


class Invoice(Printable):
    def __init__(self, order):
        self.order = order

    def print_receipt(self):
        print("Invoice:")
        for item in self.order.items:
            print(item.name, item.quantity, item.price)
        print("Total:", self.order.calculate_total())


class Order:
    def __init__(self):
        self.items = []
        self.payment = None

    def add_item(self, item):
        self.items.append(item)

    def set_payment(self, payment):
        self.payment = payment

    def calculate_total(self):
        return sum(item.total() for item in self.items)

    def checkout(self):
        total = self.calculate_total()
        self.payment.pay(total)
        invoice = Invoice(self)
        invoice.print_receipt()
```
